```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Data
species = ("Carbon", "Oxygen", "Nitrogen", "Sulfur")
penguin_means = {
    'SAM-CyNH': (18.35, 11.43, 12.98, 14),
    'SAM-CyNH + BaCBM2': (12.35, 18.43, 20.98, 10),
    'BaCBM2 DAP': (25.35, 21.43, 14.98, 18),
}

# Standard deviations for error bars
penguin_std = {
    'SAM-CyNH': (.3, .3, .1, .2),
    'SAM-CyNH + BaCBM2': (.3, .1, .4, .3),
    'BaCBM2 DAP': (.5, .45, .6, .55),
}

# Generate x positions
x = np.arange(len(species))
width = 0.25
multiplier = 0

# Apply seaborn style for nature-like aesthetics
sns.set_style("whitegrid")
sns.set_context("notebook", font_scale=1.2)

# Create the figure and axis
fig, ax = plt.subplots(figsize=(8, 4), layout='constrained')
colors = ['#1b6872', '#df208a', '#aec5fb']

# Plot the bars
for (attribute, measurement), color in zip(penguin_means.items(), colors):
    offset = width * multiplier
    std_dev = penguin_std[attribute]
    rects = ax.bar(x + offset, measurement, width, yerr=std_dev, capsize=2.2, label=attribute, color=color, edgecolor="black", alpha=0.75)
    ax.bar_label(rects, padding=3, fontsize=10)
    multiplier += 1

# Customize the axes
ax.set_ylabel('Normalized XPS Intensity', fontsize=16, fontfamily='serif', weight='bold')
ax.set_xlabel('Chemical Species', fontsize=16, fontfamily='serif', weight='bold')
ax.set_xticks(x + width, species)
ax.legend(loc='upper left', ncols=3, frameon=False, fontsize=12)
ax.set_ylim(0, 32)

# Add a grid with a natural feel
ax.grid(True, linestyle='--', linewidth=0.5, color='darkgreen', alpha=0.3)
#ax.set_axisbelow(True)

plt.show()
```
