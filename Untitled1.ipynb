```python
import ternary

# Definir os pontos para plotagem (exemplo de três composições)
points = [
    (50, 30, 20),  # 50% A, 30% B, 20% C
    (20, 40, 40),  # 20% A, 40% B, 40% C
    (30, 50, 20),  # 30% A, 50% B, 20% C
]

# Criar figura e configurar os limites
scale = 100  # Define que os valores vão de 0 a 100%
fig, tax = ternary.figure()

# Adicionar os pontos ao diagrama
tax.scatter(points, marker="o", color="red", label="Amostras")

# Configurar o grid e rótulos
tax.boundary(linewidth=2.0)  # Contorno do triângulo
tax.gridlines(multiple=10, color="gray", linestyle="dotted")  # Linhas de grade
tax.set_title("Diagrama Ternário")  # Título

# Rótulos dos eixos
tax.left_axis_label("Componente B (%)", fontsize=12)
tax.right_axis_label("Componente C (%)", fontsize=12)
tax.bottom_axis_label("Componente A (%)", fontsize=12)

# Mostrar a legenda e a figura
tax.legend()
tax.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def ternary_to_cartesian(A, B, C):
    """
    Converte coordenadas ternárias (A, B, C) em coordenadas cartesianas (x, y).
    A + B + C deve ser igual a 1 (ou 100%).
    """
    x = 0.5 * (2 * B + C) / (A + B + C)
    y = (np.sqrt(3) / 2) * C / (A + B + C)
    return x, y

# Definir pontos de exemplo no formato ternário (A, B, C)
points = np.array([
    [50, 30, 20],  # 50% A, 30% B, 20% C
    [20, 40, 40],  # 20% A, 40% B, 40% C
    [30, 50, 20],  # 30% A, 50% B, 20% C
    [10, 60, 30]   # 10% A, 60% B, 30% C
])

# Normalizar os valores para que a soma de A, B e C seja 1
points = points / 100  

# Converter os pontos para coordenadas cartesianas
cartesian_points = np.array([ternary_to_cartesian(*p) for p in points])

# Criar a figura
fig, ax = plt.subplots(figsize=(6, 6))

# Plotar os pontos
ax.scatter(cartesian_points[:, 0], cartesian_points[:, 1], color='red', label="Amostras", zorder=3)

# Definir vértices do triângulo ternário
A = np.array([0.5, np.sqrt(3)/2])  # Topo (100% C)
B = np.array([1, 0])               # Direita (100% B)
C = np.array([0, 0])               # Esquerda (100% A)

# Desenhar o triângulo
triangle = np.array([A, B, C, A])
ax.plot(triangle[:, 0], triangle[:, 1], 'k-', linewidth=2)

# Adicionar linhas de grade (divisões de 10%)
for i in range(1, 10):
    frac = i / 10
    ax.plot([frac / 2, 1 - frac / 2], [frac * np.sqrt(3) / 2, frac * np.sqrt(3) / 2], 'gray', linestyle='dotted', alpha=0.5)  # Linhas horizontais
    ax.plot([frac, frac / 2], [0, frac * np.sqrt(3) / 2], 'gray', linestyle='dotted', alpha=0.5)  # Linhas inclinadas à esquerda
    ax.plot([1 - frac, 1 - frac / 2], [0, frac * np.sqrt(3) / 2], 'gray', linestyle='dotted', alpha=0.5)  # Linhas inclinadas à direita

# Remover eixos e definir limites
ax.set_xlim(-0.1, 1.1)
ax.set_ylim(-0.1, np.sqrt(3)/2 + 0.1)
ax.set_xticks([])
ax.set_yticks([])
ax.set_frame_on(False)

# Adicionar rótulos dos eixos
ax.text(-0.05, -0.05, "A (Componente A)", fontsize=12, ha='center', va='top')
ax.text(1.05, -0.05, "B (Componente B)", fontsize=12, ha='center', va='top')
ax.text(0.5, np.sqrt(3)/2 + 0.05, "C (Componente C)", fontsize=12, ha='center', va='bottom')

# Título e legenda
ax.set_title("Diagrama Ternário - Matplotlib", fontsize=14)
ax.legend()

# Mostrar a figura
plt.show()
```
